```python
import scipy.stats as st
import numpy as np
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

Se citeste fisierul `NumeLeader.txt`

```python
Pcte=np.loadtxt('NumeLeader2.txt')
print Pcte.shape  # afiseaza c~ate linii si coloane are matricea Pcte (ar trebui n=200 linii si 2 coloane)
```

```python
X=Pcte[:, 0] # coloana 0 contine coordonata x a punctelor
Y=Pcte[:, 1]# coloana 1 contine coordonata y a punctelor
```

```python
plt.rcParams['figure.figsize'] = 12, 6
```

```python
fig = plt.figure()
ax=fig.add_subplot(111)
ax.plot(X,Y, 'bo') #deseneaza punctele
Rez= st.linregress(X,Y)# apeleaza functia care calculeaza dreapta celor mai mici patrate, y=ax+b,
                       # asociata punctelor de coordonate (X,Y)
                       # functia returneaza 5 valori: a, b, r, p sigma. Pe noi ne intereseaza primele 3
a=Rez[0]
b=Rez[1]
r=Rez[2]
ax.plot([X[0], X[-1]], [a*X[0]+b, a*X[-1]+b] , 'r', lw=2)#deseneaza dreapta cu rosu
plt.savefig('NumeLeaderDr.png')
```

La sfarsit reveniti iar in Notebook si generati distributia datelor si distributia Zipf aproximanta.

```python
pre=np.loadtxt('ProbsExp.txt')
prt=np.loadtxt('ProbsTheor.txt')
```

```python
n=200
M=max(np.max(pre), np.max(prt))
rank=range(1, n+1)
```

```python
plt.rcParams['figure.figsize'] = 12, 8
fig = plt.figure()
ax1=fig.add_subplot(211)
ax1.set_ylim([0,M])
ax1.stem(rank,  pre, 'b')#distributia datelor
ax1.set_title('Distributia datelor')
ax2=fig.add_subplot(212)
ax2.set_ylim([0,M])
ax2.stem(rank, prt, 'r', bottom=0)#distributia Zipf ce aproximeaza distributia datelor
ax2.set_title('Distributia Zipf ce aproximeaza distributia datelor')
plt.savefig('NumeLeaderZipfs.png')


    
    
```
